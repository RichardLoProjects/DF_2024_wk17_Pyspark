# Exploratory Data Analysis

Obtaining data...

```python
import pandas as pd
```

```python
file_1a = '../assets/1/train.csv'
file_1b = '../assets/1/test.csv'
file_2 = '../assets/2/Customer-Churn-Records.csv'
file_3 = '../assets/3/Bank Customer Churn Prediction.csv'
file_4 = '../assets/4/Churn Modeling.csv'
file_5 = '../assets/5/Churn_Modelling.csv'
file_6 = '../assets/6/Churn_Modelling.csv'
file_7 = '../assets/7/churn.csv'
```

```python
files = [
    file_1a
    , file_1b
    , file_2
    , file_3
    , file_4
    , file_5
    , file_6
    , file_7
]
df_lst = [pd.read_csv(file) for file in files]
print(f'df_lst contains {len(df_lst)} dataframes.')
```

Querying dataframe shape to identify whether or not it is feasible to run `df.head()`.

```python
for df_index, df_item in enumerate(df_lst):
    print(f'Dataframe {df_index} has {df_item.shape[1]} columns and {df_item.shape[0]} rows.')
```

Identifying dataframe columns...

```python
meta_df = pd.DataFrame(
    [df.columns.to_list() for df in df_lst]
    , index = [f'df{i}' for i in range(len(df_lst))]
)

meta_df
```
